```python
vgg16 = timm.create_model("vgg16", pretrained=True).to(def_device).features
normalize = transforms.Normalize(*IMAGENET_STATS)


class CalcFeatures:
    def __init__(self, target_layers):
        self.target_modules = [x for i, x in enumerate(vgg16) if i in target_layers]

    def __call__(self, im):
        if im.ndim == 3:
            im.unsqueeze_(0)

        feats = []
        def _f(h, mod, inp, outp):
            feats.append(outp.clone())

        with Hooks(self.target_modules, _f):
            vgg16(normalize(im))

        return feats



def calc_gram(feat):
    n, c, h, w = feat.shape
    flat_feats = feat.view(n, c, h*w)
    return flat_feats @ flat_feats.transpose(1, 2) / (feat.shape[-1] * feat.shape[-2])


class StyleLoss:
    def __init__(self, target_im, target_layers=(1, 6, 11, 18, 25)):
        self.target_im = target_im
        self.calc_features = CalcFeatures(target_layers)
        with torch.inference_mode():
            target_feats = self.calc_features(self.target_im)
            self.target_grams = [calc_gram(f) for f in target_feats]

    def __call__(self, input_im):
        input_feats = self.calc_features(input_im)
        input_gram_matrices = [calc_gram(f) for f in input_feats]
        return sum(
            (f1 - f2).pow(2).mean()
            for f1, f2 in zip(input_gram_matrices, self.target_grams)
        )
```

```python
def perchannel_conv(x, filters):
    """filters: [nfilters, h, w]"""
    b, ch, h, w = x.shape
    # print(x.shape)
    y = x.reshape(b*ch, 1, h, w)
    # print(y.shape)
    y = F.pad(y, [1,1,1,1], 'circular') # tiling support
    # print(y.shape)
    y = F.conv2d(y, filters[:, None])
    # print(y.shape)
    return y.reshape(b, -1, h, w)
```

```python
class SimpleCA(nn.Module):
    def __init__(self, filters, nchannels=4, nhidden=8, zero2=True):
        super().__init__()
        self.filters = filters
        nfilters = filters.size(0)

        self.brain = nn.Sequential(
            nn.Conv2d(nchannels * nfilters, nhidden, 1),
            nn.ReLU(),
            nn.Conv2d(nhidden, nchannels, 1, bias=False)
        ).to(def_device)
        if zero2:
            self.brain[2].weight.data.zero_()


    def forward(self, x, update_rate=0.5) :
        y = perchannel_conv(x, self.filters)
        y = self.brain(y)

        b, c, h, w = x.shape
        update_mask = (torch.rand(b, 1, h, w).to(x.device)+update_rate).floor()

        return x + y * update_mask

def to_rgb(x):
    return x[...,:3,:,:]+0.5
```

```python
class NCAImageLog(Callback):
    order = PlotCB.order + 1

    def __init__(self, log_every=64):
        self.log_every = log_every
    
    def before_fit(self, learn):
        self.image_fig, self.image_ax = plt.subplots(1, 1, figsize=(8, 8))

    def after_batch(self, learn):
        if learn.training and learn.iter % self.log_every == 0:
            rgb = to_rgb(learn.preds.detach()).clip(0, 1)
            im = torchvision.utils.make_grid(rgb)
            show_image(im, ax=self.image_ax, title=f"iter {learn.iter}")
            if not hasattr(self, "graph_out"):
                self.graph_out = display(self.image_fig, display_id=True)  # type: ignore
            else:
                self.graph_out.update(self.image_fig)

class NCACB(TrainCB):
    def __init__(
        self,
        style_img: torch.Tensor,
        style_loss_scale=0.1,
        size=256,
        nsteps_min=32,
        nsteps_max=96,
    ):
        self.style_img, self.style_loss_scale, self.nsteps_min, self.nsteps_max = (
            style_img,
            style_loss_scale,
            nsteps_min,
            nsteps_max,
        )
        with torch.inference_mode():
            self.pool = make_grids(256, sz=size)

    def predict(self, learn):
        # pick random samples from pool
        batch_idx = torch.randint(0, len(self.pool), (4,))
        x = self.pool[batch_idx]
        x = F.dropout2d(x, p=0.1)
        for _ in range(torch.randint(self.nsteps_min, self.nsteps_max, (1,))):
            x = learn.model(x)

        with torch.inference_mode():
            self.pool[batch_idx] = x

        learn.preds = x

    def get_loss(self, learn):
        style_loss = learn.loss_func(to_rgb(learn.preds))
        overflow_loss = (learn.preds - learn.preds.clamp(-1.0, 1.0)).abs().sum()
        learn.loss = overflow_loss + style_loss * self.style_loss_scale

    def backward(self, learn):
        learn.loss.backward()
        # gradient normalization
        for p in learn.model.parameters():
            p.grad /= p.grad.norm() + 1e-8
```
